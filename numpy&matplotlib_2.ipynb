```python
import matplotlib.pyplot as plt
import numpy as np
xlist=np.linspace(0,2*np.pi,100)
y1=np.sin(xlist)
y2=np.cos(xlist)
plt.plot(xlist,y1)
plt.plot(xlist,y2)
plt.grid()
plt.show()
```
