```python
f = open( 'wiki_인공지능.txt', 'r', encoding='UTF8')
text = f.read()
print(text)
```

```python
from PIL import Image
import numpy as np
maskHeart = np.array( Image.open('heart_mask.jpg') )
```

```python
from wordcloud import WordCloud
path =  '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
wc = WordCloud(font_path=path, width=1000, height=700, background_color='white', mask=maskHeart)
wc_output = wc.generate(text)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.imshow(wc_output)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure( figsize=(20,15) )
plt.axis('off')
plt.imshow(wc_output)
plt.show()
```
